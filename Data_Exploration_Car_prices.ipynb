# Data Exploration Analisis

```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 ## Importing our clean Dataset

```python
#Importing our clean dataset
df = pd.read_csv('cleaned_car_data.csv')
df
```

## Cheapest and most expensive car

```python
#The price of the cheapest car
df['selling_price'].min()
```

```python
#Price of the most expensive car
df['selling_price'].max()
```

```python
#The most expensive vs the cheapest car
#Cheapest
print('The cheapest car:')
df[df['selling_price'] == 359.988]
```

```python
print('The most expensive car')
df[df['selling_price'] == 120000.0]
```

## Histograms

### Counting the car brands

```python
#Number of cars of every car brand
df['name'].value_counts()
```

```python
#Count of cars by brand
sns.countplot(data=df, x='name')
plt.title("Distribution of car brands")
plt.xticks(rotation=90)
```

### Cars by type of fuel 

```python
#Count of cars by fuel
sns.countplot(data=df, x='fuel')
plt.xlabel('Type of fuel')
plt.title('Count of cars by the type of fuel')
```

### Cars by type of transmission 

```python
#Count of cars by its transmission
sns.countplot(data=df, x='transmission')
plt.xlabel('Type of transmission')
plt.title('Count pf cars by transmission')
```

### Number of owners 

```python
#Count of cars by number of owners
sns.countplot(data=df, x='owner')
plt.xlabel('Number of owners')
plt.title('Count of cars by number of owners')
```

### Cars by age

```python
#Count by age
sns.countplot(data=df, x='age')
plt.xlabel('Age of the cars (years)')
plt.title('Count of cars by the age')
plt.xticks(rotation=90)
```

## Mean price by manufacturer

```python
#Creating a table of mean price of the brands using a pivot table
prices_brand = df.pivot_table(values='selling_price', index='name', aggfunc='mean')
prices_brand = prices_brand.sort_values('selling_price', ascending = False) #Ordering from bigger to lower
prices_brand
```
