# Car Price Prediction

## Data Cleaning

```python
#Writig the libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#Showing the DataSet
df = pd.read_csv(r'C:\Datos\archive\Car details v3.csv')
df
```

```python
#Deleting any duplicate row
df.drop_duplicates()
```

```python
df = df.drop_duplicates()
df
```

We had 8128 rows and we only had 6926 left, that means we had 1202 duplicated rows.
Anyway, let´s see if there are some missing values.

```python
#Checking the type of each column
df.info()
```

```python
#Counting the missing values
df.isna().sum()
```

But first we notice that the selling price is actually given in rupees, we will convert this into dollars

```python
#Converting Rupees to dollars 
Rupees_to_dollars = 0.012
df['selling_price'] = Rupees_to_dollars*df['selling_price']
df
```

First, the Mileage column. 
We can simply drop all the missing values for this column, but instead we will replace the data, using the average mileage.
But before doing this we need to remove the signs '/kmplg'

```python
#Droping characters
df['mileage'] = df['mileage'].str.strip('/kmplg')
```

We noticed that not every cell contains numeric values in this column so we transform them into numbers

```python
df['mileage'] = pd.to_numeric(df['mileage'])
df
```

And we replace the missing values with the mean mileage

```python
avg_mil = df['mileage'].mean()
df['mileage'].fillna(value=avg_mil, inplace=True)
df
```

And we make the same procedure as we did before with the next columns

```python
df['engine'] = df['engine'].str.strip('C')
df['engine'] = pd.to_numeric(df['engine'])
df
```

```python
#Replacing missing values in engine with the mean value
mean_eng = df['engine'].mean()
df['engine'].fillna(value=mean_eng, inplace=True)
df
```

```python
df['max_power'] = df['max_power'].str.strip('bhp ') #Droping characters
df['max_power'] = pd.to_numeric(df['max_power']) #Converting to numerical 
mean_pow = df['max_power'].mean() #Calculating the mean value 
df['max_power'].fillna(value=mean_pow, inplace=True) #Replacing values
df
```

The torque column contain many characters which are dispensable, in addition to the fact that it is very difficult to keep only the values that are useful to us, so we are going to get rid of this column, so as the seller type and seats columns 

```python
df.drop(['seller_type', 'seats'], axis=1) #Droping columns
```

```python
df = df.drop(['seller_type', 'seats'], axis=1)
df
```

The names in the first column are long and complicated. Let´s just leave the brand

First we see the unique names

```python
names = df['name'].unique() 
names.sort()
names
```

And fortunately the first word of almost every row is the car´s brand, so we drop everything else

```python
df['name'] = df['name'].apply(lambda x: x.split(' ')[0]) #Droping everything after the first word
df
```

And we replace certain values whose names are the model of the car and we replace them with the brand

```python
names = df['name'].unique()
names.sort()
names
```

```python
df.replace({'name' : 
            {'Ambassador' : 'Hindustan', 'Land':'Land Rover'}  #Replacing some names for a better look
             })
```

For the year column, we will make some changes. 
Specifically we will change the year for the age of the car

First we change the year column to integers

```python
df['year'] = df['year'].astype(str).astype(np.int64) #Converting to integers
```

And we crated a new column with the age of the car

```python
Current_Year = 2023          
df['age'] = Current_Year - df['year']  #Calculating the age of the cars based on their year
df['age'] = df['age'].astype(np.int64)
df
```

So that the column year will not be necessary

```python
df = df.drop(['year'], axis=1)
```

We make a format for the fuel column since there are various types of fuel

```python
fuel_type = df['fuel'].unique()
fuel_type.sort()
fuel_type
```

We just replace CNG and LPG with gas

```python
df.replace({'fuel' : 
            {'CNG' : 'Gas', 'LPG':'Gas'}
             })             #Replacing some types of fuel for an easier look
```

```python
df = df.replace({'fuel' : 
            {'CNG' : 'Gas', 'LPG':'Gas'}
             })
df
```

```python
transmission = df['transmission'].unique()
transmission.sort()
transmission  #Types of transmission
```

We don´t have problems with the transmission, since there are only two variables

So we continue with the number of owners

We would like to replace the words with the actual number of owners. First of all we see how many types are

```python
owners = df['owner'].unique()
owners.sort()
owners #Types of owners
```

And we change the words with numbers

```python
word_to_number = {'First Owner': 1, 'Second Owner': 2, 'Third Owner':3, 'Fourth & Above Owner':4, 'Test Drive Car':1}
df.replace({'owner': word_to_number}, inplace=True)
df  #Replacing words with number for a better treatment of data
```

Finally all the columns seems right, excep for the torque, where there are still some missing  values. We simply delete all those 

```python
df.dropna()
```

```python
df = df.dropna()
df
```

```python
df.isna().sum() #Is there any missing value?
```

Now we clean the values of torque. We notice that there are more information than just the torque, so we drop all these

```python
df['torque'] = df['torque'].astype("string") #Converting to str
```

```python
df['torque'] = df['torque'].apply(lambda x: x.split(' ')[0]) #Splitting values before the space
df
```

And finally delete the signs Nm@

```python
df['torque'].str.strip('Nm@') #Deleting characters
```

```python
df['torque'] = df['torque'].str.strip('Nnmkg@')
df
```

```python
torque_b = df[df['torque'] == '380Nm(38.7kgm)'].index
torque_b   #Another value with characters
```

```python
df.drop(torque_b, inplace=True) #Droping it
```

```python
df['torque'] = df['torque'].apply(lambda x: x.split('(')[0])
df
#Droping values before the ( sign
```

```python
df['torque'] = df['torque'].str.strip('NM') #Deleting characters
```

```python
df['torque'] = df['torque'].values.astype(float) #Converting to float
```

This is our clean dataset
