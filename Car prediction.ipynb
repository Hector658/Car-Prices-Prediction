# Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X_power = df[['max_power']]
y_power = df['selling_price']
```

```python
linear_reg = LinearRegression().fit(X_power, y_power)
linear_reg
```

```python
print('Coefficient',linear_reg.coef_)
print('Intercept',linear_reg.intercept_)
```

```python
linear_reg.predict(df[['max_power']])
```

```python
X_min_max = np.array([[df['max_power'].min()], [df['max_power'].max()]])
X_min_max
```

```python
y_predict = linear_reg.predict(X_min_max)
y_predict
```

```python
sns.lmplot(data=df, x='max_power', y='selling_price', ci=None)
plt.plot(X_min_max, y_predict, "g-")
plt.xlabel('Power')
plt.ylabel('Price')
plt.show()
```
